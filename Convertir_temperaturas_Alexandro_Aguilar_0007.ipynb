### 1. Fórmulas de Conversión de Temperatura

Para realizar las conversiones entre estas dos escalas, utilizamos las siguientes relaciones matemáticas:

* **De Grados Celsius (°C) a Grados Fahrenheit (°F):**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(O de forma simplificada: $°F = (°C \times 1.8) + 32$)*

* **De Grados Fahrenheit (°F) a Grados Celsius (°C):**
    $$°C = \frac{(°F - 32) \times 5}{9}$$
    *(O de forma simplificada: $°C = \frac{°F - 32}{1.8}$)*

### 2. Ejemplos Resueltos Paso a Paso

#### Ejemplo 1: Convertir 25°C a °F (Temperatura ambiente agradable)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Realizar la multiplicación:** $25 \times 1.8 = 45$
4.  **Realizar la suma final:** $45 + 32 = 77$
* **Resultado:** 25°C equivalen a **77°F**.

#### Ejemplo 2: Convertir 98.6°F a °C (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $°C = \frac{°F - 32}{1.8}$
2.  **Sustituir el valor:** $°C = \frac{98.6 - 32}{1.8}$
3.  **Realizar la resta en el numerador:** $98.6 - 32 = 66.6$
4.  **Realizar la división final:** $\frac{66.6}{1.8} = 37$
* **Resultado:** 98.6°F equivalen a **37°C**.

# Función para convertir de Celsius a Fahrenheit

* **Funciones `ConvertirCF` y `ConvertirFC`**: Realizan el cálculo matemático. He añadido un bloque `try-except` para evitar que el programa se cierre si el usuario escribe letras en lugar de números.
* **Función `menu()`**: Solo se encarga de imprimir visualmente las opciones para mantener el código organizado.
* **`while True`**: Crea un bucle infinito que permite al usuario hacer varias conversiones sin tener que reiniciar el programa cada vez.
* **`if/elif/else`**: Controla el flujo según la opción elegida.
* **`break`**: Es la instrucción que rompe el bucle `while` y termina el programa cuando se selecciona la opción 3.

````python
# Función para convertir de Celsius a Fahrenheit
def ConvertirCF():
    celsius = float(input("Introduce los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"\n>>> {celsius}°C es igual a {fahrenheit:.2f}°F\n")

# Función para convertir de Fahrenheit a Celsius
def ConvertirFC():
    fahrenheit = float(input("Introduce los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"\n>>> {fahrenheit}°F es igual a {celsius:.2f}°C\n")

# Función para mostrar el menú de opciones
def menu():
    print("---------- MENÚ DE CONVERSIONES ----------")
    print("1. Convertir grados centígrados a Fahrenheit")
    print("2. Convertir grados Fahrenheit a grados centígrados")
    print("3. Salir")
    print("------------------------------------------")

# Bucle principal interactivo
while True:
    menu()
    opcion = input("Selecciona una opción (1-3): ")

    if opcion == "1":
        ConvertirCF()
    elif opcion == "2":
        ConvertirFC()
    elif opcion == "3":
        print("Saliendo del programa... ¡Hasta luego!")
        print("Alexandro Aguilar NL = 1")
        break
    else:
        print("\nOpción no válida. Intenta de nuevo seleccionando 1, 2 o 3.\n")
````
